## 4 Machine learning Terminology 
### 4.1 Feature and label
- Here area, bedrooms, age are called independent variables or features whereas price is a dependant variable

### 4.2 Models 
- A machine learning model is simply a rule , or a formal, which predicts a label from the feature
- In this case the model is the equation we found for the price

### 4.3 Prediction 
- This prediction is simply the output of the model


### 4.4 Formula
- Y = m1x1 + m2x2 + m3x3 + b
- price = m1 * area + m2 * bedroom + m3 * age + b

### Pickling 
# - Writing state of object info in a file 
### Unpickling
# - Read state of object info from a file


- Based on requirement we can write total state of object to the file and we can read total object infomation from the file
## 1. Pickling
- The Process of writing state of object to the file is called pickling
- We can implement pickling and unpickling by using python pickle module
- We can do pickling by using dump ( object,file ) predefined function in pickle module 

# - pickle.dump(object,file)

## 2 Unpickling
- The process of reading state of an object from the file is called unpickling
- We can do unpickling by using load(file) predefined function in pickle module



# - obj = pickle.load(file)

# 1 dump(p1,p2) => We can do pickling
# 2 load(p) => we can do unpickling 

```python
# pickling and unpickling

import pickle

class Employee:
    def __init__(self,eno,ename,esal):
        self.eno=eno
        self.ename=ename
        self.esal=esal
    def display(self):
        print("Number is :",self.eno)
        print("Name is :",self.ename)
        print("Salary is :",self.esal)

#e1=Employee(101,"kishan",100000.49)
#e1.display()

# open(new of file , mode of file)
with open("emp.dat","wb") as f:  # em.dat is a file name and wb is writing in binary from
    e=Employee(101,"kishan",100000.49)
    pickle.dump(e,f)
    print("Pickling of employee object completed...\n")

with open("emp.dat","rb") as f:   # rb is read  binary
    obj=pickle.load(f)
    print("pickling of employee infomation after unpickling")
    obj.display()

    
```

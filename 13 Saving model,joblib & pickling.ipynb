```python
# Can i save reg object ..?
# yes we can save by using pickling topic 

import pandas as pd
import pickle
from sklearn.linear_model import LinearRegression

df=pd.read_csv("/Users/apple/Downloads/area_price_table.csv")
#print(df)
new_df=df.drop("price",axis="columns")

reg=LinearRegression()

reg.fit(new_df.values,df.price.values)


#prediction
print(reg.predict([[5000]]))
print()


with open("predict_area","wb") as f:
    pickle.dump(reg,f)
    print("req object we saved into model.predict_area")
    print("pickling is done ")

with open("predict_area","rb") as f:
    reg=pickle.load(f)
    print("pickling loading is done")
    print(reg.predict([[5000]]))
```

## using Joblib

```python
# Save trained model using joblib

import pandas as pd
import joblib
from sklearn.linear_model import LinearRegression

df=pd.read_csv("/Users/apple/Downloads/area_price_table.csv")

new_df=df.drop("price",axis="columns")

model=LinearRegression()

model.fit(new_df.values,df.price.values)




joblib.dump(model,"model_joblib")
print("Model got saves")
```

```python
# Do prediction with saved model

import joblib

obj=joblib.load("Model_joblib")
print(obj.predict([[5000]]))
```

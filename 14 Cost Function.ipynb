## 1 Cost Function in Machine learning
- Cost function are also called as loss functions in machone learning
- These are optimization techniques(Machine Learning to improve model performance by minimizing error (loss) or maximizing accuracy.)
- There are many cost function in machine learning
- There are different cost function for regression and classication

## 2 Why Cost Function in Machine Learning
- During the training phase , the model assumes the initial weights randomly and tries to make a prediction on training data.
- So here how the model gets to know how much "far" it was from the prediction?
- Model needs this infomation so that it can adjust the weight accodingly (using gradient descent) in the next iteration on training

## 3 How cost function works ?
- Cost Function is a function that takes both predicted outputs and actual outputs from model.
- Then it calculates how much error in the model to predict the good results
- In next step the model tries to adjust this error for the next iteration on training data to reach optimization level

## 4 Goal of training 
- The goal of the training phase is to come up with weights that minimize the error in every iteration
- This way of training is required in machine learning to optimize model

```python
# step 1 Importing Libraries
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics

# step 2 Loading Dataset 
df=pd.read_csv("/Users/apple/Downloads/hours_scores.csv")
# print(df)

# step 3 Data prepartion 
X=df.iloc[:,:-1].values
#print(X)
y=df.iloc[:,1].values
#print(y)


# step 4 Splitting the dataset
X_train,X_test,y_train,y_test=train_test_split(X,
                                               y,
                                               test_size=0.2,
                                               random_state=0)

# step 5 model creation 
model=LinearRegression()


# step 6 training model
model.fit(X_train,y_train)

#print(y_test)
# step 7 model  prediction
y_pred=model.predict(X_test)
print(y_pred)
```

```python
print("Score:=", model.score(X_train,y_train))


mae=metrics.mean_absolute_error(y_test,y_pred)
mse=metrics.mean_squared_error(y_test,y_pred)
rmse=metrics.root_mean_squared_error(y_test,y_pred)

print("Mean Absolute Error :=",mae)
print("Mean squared Error :=",mse)
print("Root Mean squared Error :=",rmse)
```

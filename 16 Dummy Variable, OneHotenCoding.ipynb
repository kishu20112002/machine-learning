```python
# step 5 model creation


import pandas as pd 
from sklearn.linear_model import LinearRegression

df=pd.read_csv("/Users/apple/Downloads/homeprices.csv")

dummies=pd.get_dummies(df.town,dtype=int)
merge=pd.concat([df,dummies],axis="columns")
print(merge)
print()

final=merge.drop("town",axis="columns")
print(final)

X=final.drop("price",axis="columns").values
print()
print(X)
y=final.price.values

model=LinearRegression()

# step 6 model training

model.fit(X,y)

# step 7 prediction 
result=model.predict([[2400,1,0,0]])
print()
print(result)
```

```python
# Checking the score   the training scorce of 70%

import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

df=pd.read_csv("/Users/apple/Downloads/homeprices.csv")

dummies=pd.get_dummies(df.town, dtype=int)
#print(dummies)
merge=pd.concat([df,dummies],axis="columns")
print()
#print(merge)

final=merge.drop("town",axis="columns")
print()
#print(final)

X=final.drop("price",axis="columns").values
print()
#print(X)
y=final.price.values
print()
#print(y)

model=LinearRegression()

model.fit(X,y)

print()
print(model.score(X,y))


```

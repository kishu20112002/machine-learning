## 1 Gradient DEscent 
# - Gradientt descent is an optimization algorithm
- This algorithm find the values of paramenters (coeffcients) of a function (f) that minimizes a cost function (cost)


## 2 How Gradient Descent works
1. Start with random initial values for the parameters
2. Predict the value of the target variable using the current parameters
3. Calculate the cost associated with the prediction
4. Is cost minimized ?
   - IF yes, then got step no - 6
   - If No, then go to step - 5
5. Update to parameter values using the gradient descent algorthim and return to step no -2
6. We have our final updated parameters
7. Our model is ready

| **Step No.** | **Stage**           | **Description**                                                   |
| ------------ | ------------------- | ----------------------------------------------------------------- |
| 1            | Parameters          | Initialize model parameters (weights / coefficients)              |
| 2            | Prediction          | Predict output using current parameters                           |
| 3            | Error               | Calculate the error (difference between actual & predicted value) |
| 4            | Check Error         | Check whether the error is **small or big**                       |
| 5            | If Error is Small   | Stop training (model has converged)                               |
| 6            | If Error is Big     | Update the parameters                                             |
| 7            | Update Parameters   | Adjust parameters using Gradient Descent                          |
| 8            | Prediction (Repeat) | Make a new prediction with updated parameters                     |
| 9            | Error (Repeat)      | Recalculate error                                                 |
| 10           | Loop                | Repeat steps until error becomes small                            |

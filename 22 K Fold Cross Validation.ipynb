```python
# K fold cross validation 3 in one

from sklearn.model_selection import cross_val_score
from sklearn.datasets import load_digits
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC 
from sklearn.ensemble import RandomForestClassifier


digits=load_digits()

model1=LogisticRegression(solver="lbfgs",max_iter=5000)
model2=SVC()
model3=RandomForestClassifier(n_estimators=40)


scorc1=cross_val_score(model1,digits.data,digits.target,cv=3)
scorc2=cross_val_score(model2,digits.data,digits.target,cv=3)
scorc3=cross_val_score(model3,digits.data,digits.target,cv=3)


print("Checking all score")
print()
print("1 LogisticRegression ")
print(scorc1)
print()
print("2 SVM")
print(scorc2)
print()
print("3 Random Forest")
print(scorc3)



```

```python
# K fold cross validation 3 in one

import numpy as np
from sklearn.model_selection import cross_val_score
from sklearn.datasets import load_digits
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC 
from sklearn.ensemble import RandomForestClassifier


digits=load_digits()

model1=LogisticRegression(solver="lbfgs",max_iter=5000)
model2=SVC()
model3=RandomForestClassifier(n_estimators=40)


scorc1=cross_val_score(model1,digits.data,digits.target,cv=3)
scorc2=cross_val_score(model2,digits.data,digits.target,cv=3)
scorc3=cross_val_score(model3,digits.data,digits.target,cv=3)




print("average all score")
print()
print("1 LogisticRegression ")
print(np.average(scorc1)*100)
print
print()
print("2 SVM")
print(np.average(scorc2)*100)

print()
print("3 Random Forest")
print(np.average(scorc3)*100)


```

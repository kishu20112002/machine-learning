## 4 The goal of supervised algorithm
- In supervised learning technique, an algorithm learn a model from training data

  - Find the best mapping function / target function (f) for the output variable (Y) for the given input data(X)
 
## 5 Error 
- Error means some wrong calculation.
- Error can be small error or big error
- Small error can be acceptable and big error should be minimize

## 6 Prediction error
- Prediction error means we made some prediction but there are some error are existing
- The prediction error for any machine learning can be divided into two parts
  - Bias Error
    - Low bias
    - High bias
  - Variance Error
    - Low variance 
    - High variance
## 7 Bias Error 
- Bias are the simplifying assumptions made by to make the target function easy to learn

## 8 Low Bias
- Low bias value suggests more assumptions about the from of the target function

## 9 High-Bias
- High Bias valuse suggests low assumptions about the from of the target function

## 10 Example

## 11 Variance Error 
- Variance is the error that occurs due to small changes in the training set.
- The target function is estimated from the training data by a machine learning algorithm , we can expect the algorithm may have some variance.
- Ideally it should not change too much from one training dataset to the other
- 70 + 10 small change
- 70 + 120 big change

## 12 Low Variance
- Low variance value suggests small changes to the estimate of the target function with changes to the training dataset

## 13 High variance
- High variance value suggests large changes to the estimate of the target function with changes to the training datastet

## 14 Example 
- Low - variance machine learning algorithms are
  - Linear Regression
  - Linear Discriminant Analysis
  - Logistic Regression
    
- High - variance machine learning algorithms are,
  - Decision Trees
  - k-Nearest Neighbors
  - Support Vector Machines

## 15 Bias - Variance Trade-Off
- The goal of any supervised machine learning algorithm is to achieve the sweet spot of low bias and low variance
- It means the algorithm should achieve good prediction performance.

# note
- Parametric or linear machine learning algorithm ofter have a high bias but a low variance
- Nonparametric or nonlinear machine learning algorithm often have a low bias but a high variance

## 16 Bias and Variance
### Low bias and low variance

- if low bias and low variance then the error is very less
- so, in this scenario the model is very accurate

### Low bias and high variance
- So, In this scenario the model having lower accuracy

### High bias and low variance

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from mlxtend.evaluate import bias_variance_decomp

df=pd.read_csv("/Users/apple/Downloads/student_scores_26_rows.csv")

X=df.iloc[:,:-1].values
y=df.iloc[:,1].values

#print(X)
#print(y)

X_train,X_test,y_train,y_test=train_test_split(X,
                                               y,
                                              test_size=0.2,
                                              random_state=1)

model=LinearRegression()

mse,bias,var=bias_variance_decomp(model
                                  ,X_train,
                                  y_train, 
                                  X_test,
                                  y_test, 
                                  loss='mse',
                                  num_rounds=200,
                                  random_seed=1)

print("MSE:",mse)
print("Bias:",bias)
print("Variance",var)
```

## 1 Regression 
- By using regression we can find the relationship between variables.
- Once we understand the relationship in between the variable then we can predict the future outcomes

## 2 A line 
- If two variable having relationship then if we draw this relationship in a two dimension then we get a straight line
- The picture of linear regression is simple
- Let us say we have some points , a line will travel in between these points

## 3 The goal 
- The goal of linear regression is to draw the best fitted line
- Best fitted line means that line which passes as close possible to these points 

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

df = pd.read_csv("/Users/apple/Downloads/area_price_table.csv")

new_df=df.drop("price",axis="columns")

plt.xlabel("area")
plt.ylabel("price")

plt.scatter(df.area.values,
           df.price.values,
           marker="*",
           color="Red")

model=LinearRegression()
model.fit(new_df.values,df.price.values)
plt.plot(df.area.values,
        model.predict(new_df.values),
        color="Blue")
plt.show()
```

# 4 Can we use regression in everywhere  veryyy imp ques in interview
- If there is a relationship in between the variables then only we can use linear regression algorithm.
- If there is no relationship in between the variable then we cannot use linear regression algorithm

## 5 r value
- r value explains about how variables are related each other
- This is very importtant step to recognise in between values of x-axis and y-axis

## 6 r value range
- The r values range from -1 to 1.
- While calculating if we get r value as near to -1 or 1 then those variables are strongly related each other.
- While calculating if we get r value as 0 then it confirms that there is no relationship in between the variables.

## 7 Calculate r value
- By using python scipy model we can calculate r value easily

```python
# Plotting x and y values

import pandas as pd
#from scipy import stats
import matplotlib.pyplot as plt

d={
    "area":[1,2,3,4],
    "rice_yield":[10,20,30,40]
}

df=pd.DataFrame(d)

X=df.area.values
y=df.rice_yield.values

plt.xlabel("X")
plt.ylabel("y")

plt.scatter(X,y,marker="*")
plt.show()
```

```python
# WE can know how much relation there have 

import pandas as pd
#from scipy import stats
import matplotlib.pyplot as plt

d={
    "area":[1,2,3,4],
    "rice_yield":[10,20,30,40]
}

df=pd.DataFrame(d)

print(df.corr())
```

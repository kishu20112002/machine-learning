## 8 Evaluating the Algorithm
- We need to evaluate the performance of algorithm
- The step is particularly important to compare how well different algoithms perform on a particular dataset
- For regression algorithms there are three evaluation metrics are commonly used

## 9 Evaluating metrics   also know as cost functions
- Mean Absolute Error (MAE)
- Mean Squared Error (MSE)
- Root Mean Squared Error (RMSE)


### Mean Absolute Error (MAE)
- Mean Absolute Error (MAE) is the mean of the absolute value of the errors

### Mean Squared Error (MSE)
- Mean Squared Error (MSE) is the mean of the squared errors.

### Root Mean Squared Error (RMSE)
- Root Mean Squared Error (RMSE) is the square root of the mean of the squared errors

## WE no need to calculate manually
- We donot have to perform these calculation manually.
- The scikit-learn library comes with pre=built function that can be used to find out these values for us

```python
# step 1 Importing Libraries
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics

# step 2 Loading Dataset 
df=pd.read_csv("/Users/apple/Downloads/hours_scores.csv")
# print(df)

# step 3 Data prepartion 
X=df.iloc[:,:-1].values
#print(X)
y=df.iloc[:,1].values
#print(y)


# step 4 Splitting the dataset
X_train,X_test,y_train,y_test=train_test_split(X,
                                               y,
                                               test_size=0.2,
                                               random_state=0)

# step 5 model creation 
model=LinearRegression()


# step 6 training model
model.fit(X_train,y_train)


#print(y_test)
# step 7 model  prediction
y_pred=model.predict(X_test)
#print(y_pred)
print("Score:=", model.score(X_train,y_train))


mae=metrics.mean_absolute_error(y_test,y_pred)
mse=metrics.mean_squared_error(y_test,y_pred)
rmse=metrics.root_mean_squared_error(y_test,y_pred)

print("Mean Absolute Error :=",mae)
print("Mean squared Error :=",mse)
print("Root Mean squared Error :=",rmse)
```
